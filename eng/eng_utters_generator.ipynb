```python
from jsgf import parse_grammar_string
```

```python
with open('jsgf_eng_extended_rules.txt', 'r') as file:
    strings = file.read()
```

```python
try:
    grammar = parse_grammar_string(strings)
except:
    print("A parsing error: Unable to process the grammar file")
```

```python
print("The names of rules are in" + 'jsgf_eng_extended_rules.txt' + 'as follows:')
for count, rule in enumerate(grammar.rules):
    print(count, rule.name)
```

```python
gen_utters = []
for i in range(1000):
    gen_utters.append(grammar.rules[0].generate())
print(*set(gen_utters), sep='\n')
```

```python
with open('jsgf_eng_extended_rules.txt', 'r') as file:
    strings = file.read()
```

```python
with open('jsgf_eng_extended_rules' + "_generated_utterances.txt", 'w') as f:
    for utter in gen_utters:
        f.write("%s\n" % utter)
```
